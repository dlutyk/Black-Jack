```python
import random

kolory = ("Kier", "Karo", "Pik", "Trefl")
rangi = ("dwójka", "trójka", "czwórka", "piątka", "szóstka", "siódemka", "ósemka", "dziewiątka", "dziesiątka", 
         "walet", "królowa", "król", "as")
wartości = {"dwójka":2, "trójka":3, "czwórka":4, "piątka":5, "szóstka":6, "siódemka":7, "ósemka":8, "dziewiątka":9, 
            "dziesiątka":10, "walet":10, "królowa":10, "król":10, "as":11}
playing = True

class Card:
    def __init__(self, kolor, ranga):
        self.kolor = kolor
        self.ranga = ranga
        
    def __str__(self):
        return self.ranga + " " + self.kolor

class Deck:
    def __init__(self):
        self.talia = []
        for kolor in kolory:
            for ranga in rangi:
                self.talia.append(Card(kolor, ranga))
    
    def tasuj(self):
        random.shuffle(self.talia)
        
    def rozdaj(self):
        poj_karta = self.talia.pop()
        return poj_karta


class Hand:
    def __init__(self):
        self.karty = []
        self.suma = 0
        self.asy = 0
        
    def add_card(self, karta):
        self.karty.append(karta)
        self.suma += wartości[karta.ranga]
        if karta.ranga == "as":
            self.asy += 1
    
    def adjust_for_ace(self):
        while self.suma > 21 and self.asy >= 1:
            self.suma -= 10
            self.asy -= 1

class Chips:
    
    def __init__(self):
        self.total = 100
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet


def take_bet(kasa):
    while True:
        try:
            kasa.bet = int(input("\nile stawiasz?: "))
        except:
            print("\nto nie jest liczba gapo")
        else:
            if kasa.bet > kasa.total:
                print("\nnie masz tyle! Idź do roboty!")
            else:
                break

def hit(deck,hand):
    hand.add_card(deck.rozdaj())
    hand.adjust_for_ace()
        
def hit_or_stand(deck,hand):
    global playing
    chcesz = input("\nchcesz kolejną kartę? Wpisz tak lub nie: ").lower()
    while True:
        if chcesz == "tak":
            hit(deck,hand)
        elif chcesz == "nie":
            playing = False
        else:
            continue
        break

def show_some(player,dealer):
    print("\nkarty krupiera: \n%s\n%s" %("ukryta karta", dealer.karty[1]))
    print("\nkarty gracza: ", *player.karty, sep = "\n")
    print("punkty gracza: %s" %(player.suma))

def show_all(player,dealer):
    print("\nkarty krupiera: ", *dealer.karty, sep = "\n")
    print("punkty krupiera: %s" %(dealer.suma))
    print("\nkarty gracza: ", *player.karty, sep = "\n")
    print("punkty gracza: %s" %(player.suma))
    

def player_busts():
    print("\nTwój wynik przekracza 21")
    print("PRZEGRYWASZ")
    kasa.lose_bet()
    print("twoja kasa: %s" %(kasa.total))

def player_wins():
    print("\nBRAWO! WYGRYWASZ!")
    kasa.win_bet()
    print("twoja kasa: %s" %(kasa.total))
    
def dealer_busts():
    print("\nWynik krupiera przekracza 21")
    print("BRAWO! WYGRYWASZ!")
    kasa.win_bet()
    print("twoja kasa: %s" %(kasa.total))
    
def dealer_wins():
    print("\nPRZEGRYWASZ")
    kasa.lose_bet()
    print("twoja kasa: %s" %(kasa.total))

def remis():
    print("\nREMIS")
    print("twoja kasa: %s" %(kasa.total))

kasa = Chips()

print("SIEMKA! Gramy w Black Jack.\n\
Chodzi o to, żeby suma kart była jak najbliższa 21 ale nie przekroczyła tej liczby!\n\
Krupier dobiera karty póki ich suma nie przekroczy 17.\n\
As liczy się jako 1 lub 11")

while True:
    print("\nNOWA GRA")
    deck = Deck()
    deck.tasuj()
    
    player = Hand()
    hit(deck,player)
    hit(deck,player)
    
    dealer = Hand()
    hit(deck,dealer)
    hit(deck,dealer)
        
    print("Twoja kasa: %s" %(kasa.total))
    
    take_bet(kasa)
    
    show_some(player,dealer)
    
    while playing:
        hit_or_stand(deck, player)
        show_some(player,dealer)
        
        if player.suma > 21:
            player_busts()
            break
        
    if player.suma <= 21:
        while dealer.suma < 17:
            hit(deck,dealer)
        
        show_all(player,dealer)
    
        if dealer.suma > 21:
            dealer_busts()
            
        elif player.suma > dealer.suma:
            player_wins()

        elif player.suma < dealer.suma:
            dealer_wins() 

        else:
            remis()

    if input("\nchcesz zagrać jeszcze raz? Wpisz tak lub nie: ").lower() == "tak":
        playing = True
        continue
    else:
        print("No to koniec zabawy :(")
        break
```
